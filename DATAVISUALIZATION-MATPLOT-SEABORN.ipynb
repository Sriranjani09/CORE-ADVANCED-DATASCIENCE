***Import Libraries*** 

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

***Creating Sample data's***

```python
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [23, 45, 56, 78],
    'X': [5, 7, 8, 7],
    'Y': [7, 8, 5, 6]
}
df = pd.DataFrame(data)
```

***Matplotlib (Plot)***

```python
# Line Plot
plt.figure(figsize=(6,4))
plt.plot(df['Category'], df['Values'], marker='o')
plt.title('Matplotlib - Line Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python

# Bar Chart
plt.bar(df['Category'], df['Values'], color='orange')
plt.title('Matplotlib - Bar Chart')
plt.show()


```

```python
# Scatter Plot
plt.scatter(df['X'], df['Y'], color='green')
plt.title('Matplotlib - Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
                                                                                                                              
```

```python

# Histogram
plt.hist(df['Values'], bins=5, color='purple')
plt.title('Matplotlib - Histogram')
plt.show()
```

```python

# Pie Chart
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%')
plt.title('Matplotlib - Pie Chart')
plt.show()
```

***Seaborn(Plot)***

```python
# Set style
sns.set(style="whitegrid")

# Line Plot
sns.lineplot(x='Category', y='Values', data=df, marker='o')
plt.title('Seaborn - Line Plot')
plt.show()

```

```python
# Bar Plot
sns.barplot(x='Category', y='Values', data=df, palette='viridis')
plt.title('Seaborn - Bar Plot')
plt.show()

```

```python
# Scatter Plot
sns.scatterplot(x='X', y='Y', data=df, color='red', s=100)
plt.title('Seaborn - Scatter Plot')
plt.show()

```

```python
# Histogram
sns.histplot(df['Values'], bins=5, color='skyblue', kde=True)
plt.title('Seaborn - Histogram with KDE')
plt.show()
```
